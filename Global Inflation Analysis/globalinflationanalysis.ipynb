**Global Inflation Analysis and Visualisations**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Import dataset, Global Inflation from 1980-2024
df = pd.read_csv("global_inflation_data.csv")
```

```python
# US and China
china_inf = df[df["country_name"] == "China, People's Republic of"]
us_inf = df[df["country_name"] == "United States"]

china_data = china_inf.drop(columns=["country_name", "indicator_name"])
us_data = us_inf.drop(columns=["country_name", "indicator_name"])
```

```python
china_data = china_data.transpose().reset_index()
us_data = us_data.transpose().reset_index()
```

```python
china_data.columns = ["Year", "Inflation"]
us_data.columns = ["Year", "Inflation"]
```

```python
china_data["Year"] = china_data["Year"].astype(int)
us_data["Year"] = us_data["Year"].astype(int)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(china_data["Year"], china_data["Inflation"], marker="o", color='blue', label="China Inflation")
plt.plot(us_data["Year"], us_data["Inflation"], marker="o", color="orange", label="US Inflation")
plt.title("Average Yearly Inflation (1980-2024)")
plt.xlabel("Year")
plt.ylabel("Inflation Rate (%)")
plt.legend(loc='best')
plt.grid(True)

plt.show()
```

The inflation trajectories of China and the United States from 1980 to 2024 reflect fundamentally different economic evolutions. In the 1980s and early 1990s, China underwent periods of high inflation (often exceeding 15%) as the country transitioned from a centrally planned economy to a market-oriented one. These spikes were largely driven by price liberalizations and rapid economic reforms.

In contrast, the United States maintained a more stable inflation environment over the same period, with only occasional fluctuations, such as the early 1980s inflation surge due to oil shocks, and a moderate increase following the 2008 financial crisis. U.S. inflation remained generally controlled through consistent Federal Reserve policy and a mature market economy.

Since the 2000s, China has managed to stabilize its inflation, often keeping it between 1% and 4%, signaling improved macroeconomic control. However, the pandemic years (2020–2022) introduced a global inflation shock, during which both countries saw modest to sharp increases though the U.S. experienced a more visible post-pandemic surge, especially in 2022.

This comparison highlights how differing stages of development, central bank roles, and policy responses shape long-term inflation dynamics in two of the world’s largest economies.

```python
#UK and US
uk_inf = df[df['country_name'] == 'United Kingdom']
us_inf = df[df['country_name'] == 'United States']

uk_data = uk_inf.drop(columns=['country_name', 'indicator_name'])
us_data = us_inf.drop(columns=['country_name', 'indicator_name'])


uk_data = uk_data.transpose().reset_index()
us_data = us_data.transpose().reset_index()


uk_data.columns = ['Year', 'Inflation']
us_data.columns = ['Year', 'Inflation']


uk_data['Year'] = uk_data['Year'].astype(int)
us_data['Year'] = us_data['Year'].astype(int)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(uk_data['Year'], uk_data['Inflation'], marker='o', color='blue', label='UK Inflation')
plt.plot(us_data['Year'], us_data['Inflation'], marker='o', color='red', label='US Inflation')
plt.title('Average Yearly Inflation (1980-2024)')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.legend(loc='best')
plt.grid(True)

plt.show()
```

Over the last four decades, the inflation patterns of the United Kingdom and the United States have shared similarities but also diverged at key moments, reflecting differing fiscal environments and monetary policies.

In the 1980s, both countries grappled with high inflation (particularly in the early years) prompting strong interest rate responses from central banks. The UK experienced steeper inflation rates during parts of the 1980s and early 1990s, driven in part by volatile monetary policy transitions and the effects of entering and later exiting the European Exchange Rate Mechanism (ERM).

The US, on the other hand, maintained a comparatively more stable inflation path throughout the 1990s and 2000s, benefiting from a robust Federal Reserve policy and steady economic growth.

Post-2008 financial crisis, both countries saw subdued inflation levels for years. However, the COVID-19 pandemic marked a turning point. From 2021 onward, both nations experienced a sharp rise in inflation, with the US witnessing a slightly more aggressive spike, largely due to higher fiscal stimulus and supply chain disruptions.

Overall, while both economies show long-term alignment in inflation control, periods of divergence highlight the impact of country-specific policy choices and structural differences.

```python
# Germany and Türkiye

de_inf = df[df['country_name'] == 'Germany']
tr_inf = df[df['country_name'] == 'Türkiye, Republic of']

de_data = de_inf.drop(columns=['country_name', 'indicator_name'])
tr_data = tr_inf.drop(columns=['country_name', 'indicator_name'])

de_data = de_data.transpose().reset_index()
tr_data = tr_data.transpose().reset_index()

de_data.columns = ['Year', 'Inflation']
tr_data.columns = ['Year', 'Inflation']

de_data['Year'] = de_data['Year'].astype(int)
tr_data['Year'] = tr_data['Year'].astype(int)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(de_data['Year'], de_data['Inflation'], marker='o', color='blue', label='DE Inflation')
plt.plot(tr_data['Year'], tr_data['Inflation'], marker='o', color='red', label='TR Inflation')
plt.title('Average Yearly Inflation (1980-2024)')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.legend(loc='best')
plt.grid(True)

plt.show()
```

This comparison of inflation rates between Germany and Turkey from 1980 to 2024 reveals striking differences in economic stability and monetary policy. While Germany has maintained relatively low and stable inflation throughout the decades (typically under 5%),Turkey has experienced significant volatility. Particularly during the 1990s and early 2000s, Turkey faced periods of hyperinflation, with annual rates exceeding 60%, reflecting structural economic challenges and policy shifts.

In the post-2010 era, while Germany continued on a path of moderate inflation, Turkey again saw a sharp upward trend, especially after 2018, reaching double-digit inflation that persisted into the 2020s. The contrast becomes even more pronounced during and after the COVID-19 pandemic, highlighting differences in fiscal discipline, central bank independence, and macroeconomic management.

This long-term comparison underscores how sustained economic policies and institutional trust can shape a country’s inflation profile across decades.

```python
#Russia and US
rs_inf = df[df['country_name'] == 'Russian Federation']
us_inf = df[df['country_name'] == "United States"]

rs_data = rs_inf.drop(columns=['country_name', 'indicator_name'])
us_data = us_inf.drop(columns=['country_name', 'indicator_name'])

rs_data = rs_data.transpose().reset_index()
us_data = us_data.transpose().reset_index()

rs_data.columns = ['Year', 'Inflation']
us_data.columns = ['Year', 'Inflation']

rs_data['Year'] = rs_data['Year'].astype(int)
us_data['Year'] = us_data['Year'].astype(int)
```

```python
rs_filtered = rs_data[rs_data['Inflation'] < 100]
us_filtered = us_data[us_data['Inflation'] < 100]

plt.figure(figsize=(10, 6))
plt.plot(rs_filtered['Year'], rs_filtered['Inflation'], marker='o', color='blue', label='Russia Inflation')
plt.plot(us_filtered['Year'], us_filtered['Inflation'], marker='o', color='red', label='US Inflation')
plt.title('Yearly Inflation ')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.legend(loc='best')
plt.grid(True)

plt.show()
```

When comparing inflation rates between Russia and the United States, distinct patterns emerge that reflect differences in economic transitions and policy responses. Following the dissolution of the Soviet Union in the early 1990s, Russia experienced a period of hyperinflation and economic instability, while the United States maintained relatively stable inflation during the same period. Throughout the 2000s and beyond, Russia’s inflation showed greater variability, whereas the U.S. inflation trend remained more moderate. Both countries experienced inflationary pressure during global economic shocks, such as the 2008 financial crisis and the COVID-19 pandemic. In recent years, monetary and fiscal responses to these events have led to inflation increases in both countries, though the magnitude and timing of these changes have differed.

```python
#China annual change

china_data['Inflation Change'] = china_data['Inflation'].diff()

# Drop the first row with NaN value
china_data = china_data.dropna()

colors = ['red' if x > 0 else 'green' for x in china_data['Inflation Change']]

plt.figure(figsize=(10, 6))
plt.bar(china_data['Year'], china_data['Inflation Change'], color=colors)
plt.title('China Annual Change in Inflation (1980-2024)')
plt.xlabel('Year')
plt.ylabel('Change in Inflation Rate (%)')
plt.grid(True)
plt.show()
```

```python
#US annual change

us_data['Inflation Change'] = us_data['Inflation'].diff()

us_data = us_data.dropna()

colors = ['red' if x > 0 else 'green' for x in us_data['Inflation Change']]

# Plot the annual change
plt.figure(figsize=(10, 6))
plt.bar(us_data['Year'], us_data['Inflation Change'], color=colors)
plt.title('US Annual Change in Inflation (1980-2024)')
plt.xlabel('Year')
plt.ylabel('Change in Inflation Rate (%)')
plt.grid(True)
plt.show()
```

Between 2000 and 2024, inflation patterns varied noticeably among the US, UK, China, and Russia. The US and UK generally followed a stable path, with spikes during global events like the 2008 financial crisis and especially after the COVID-19 pandemic. China kept inflation relatively low and steady, likely due to tight government control and slower price fluctuations. Russia’s inflation was more volatile, reflecting the impact of political shifts, sanctions, and economic restructuring. While all four countries felt the effects of global inflationary pressure in the 2020s, how they responded and how quickly,differed significantly.
