```python
list_features = ['Social support','Freedom to make life choices','Perceptions of corruption'] #distribution of Social support,Freedom to make life choices,Perceptions of corruption
sns.boxplot(data=df24.loc[:,list_features],orient='h',palette = 'Set3')
plt.show()
```

**Happiest and Unhappiest Countries in 2024**

```python
df24_happiest_unhappiest = df24[(df24.loc[:,'Ladder score']>7.4)|(df24.loc[:,'Ladder score']<3.5)]
sns.barplot(x = 'Ladder score', y= 'Country name',data = df24_happiest_unhappiest, palette = 'coolwarm')
plt.title('Happiest and Unhappiest Countries in 2024')
plt.show()
```

In this section, we constructed a chart that shows the happiest and unhappiest countries based on the Ladder Score value. When setting the boundaries, we referred to the chart created in the previous step. We established two limits, one for the lower and one for the upper end, based on the ladder score graph.

If we were to draw a conclusion here, we could say that Western European countries tend to be happier than other nations. In the next section, we will re-evaluate this conclusion by conducting a region-based ranking to check its accuracy.

```python
plt.figure(figsize = (15,8))
sns.kdeplot(x=df24["Ladder score"], hue = df24["Regional indicator"], fill = True, linewidth = 2)
plt.axvline(df24["Ladder score"].mean(),c= "black")
plt.title("Distribution of Ladder Scores by Regional Indicator")
plt.show()
```

In this section, we compared the ladder score values on a regional basis using a different visualization format.

According to the chart, we can conclude that happiness is low in Africa and South Asia, while the happiness rate is higher in Western Europe. This conclusion also supports the findings we derived from the previous graph.

In this section, a choropleth map was created showing the 'Life Ladder' values over the years. Additionally, an animation has been included that changes over time. The "year" feature was used to establish the time-based relationship.

The color scale on the right side of the map indicates the grading of life ladder values. By referring to this scale, we can determine the life ladder value of each country based on the color assigned to it on the map.

```python
import plotly.express as px

worldmap = px.choropleth(
    df.sort_values("year"), 
    locations="Country name",
    color="Life Ladder",
    locationmode="country names",
    animation_frame="year",
    color_continuous_scale="RdYlGn"
)

worldmap.update_layout(title="Life Ladder Comparison by Countries")
worldmap.show()
```

```python
df24freedom = df24[(df24.loc[:,"Freedom to make life choices"]>0.830)|(df24.loc[:,"Freedom to make life choices"]<0.30)].sort_values("Freedom to make life choices", ascending=False)
sns.barplot(x = "Freedom to make life choices", y= "Country name",data = df24freedom, palette = "coolwarm")
plt.title("Most and Least Free Countries to Make Life Choices in 2024")
plt.show()
```

While Cambodia ranks the highest in terms of Freedom to Make Life Choices according to this chart, it does not rank among the top countries in the World Happiness Report 2024. Despite its strong performance in personal freedom, Cambodia's low Ladder Score—which factors in indicators like Social Support, Healthy Life Expectancy, Generosity, Perceptions of Corruption etc.highlights that freedom alone does not guarantee national happiness.
